## One dimensional Gradient descent

We first understand the function given, that is f(x) = -x**2

We assign a variable x, with 100 random values between -50 to 50. Then we make a function f(x) as name it y which gives us corresponding function values for x.

```python
x = np.random.randint(low = -50, high = 50, size = 100)
print(x)
y = -1*(x**2)
print(y)
```

We plot the function using matplotlib. We need to find the maxima of this function using gradient ascent.

```python
fig, ax = plt.subplots()
ax.scatter(x,y)
```

Now we make a 1 dimensional function and ascent over the function value.

```python
# we initialze a function which accepts learning rate and one value of x
def grad_asc(R, x_0):
    # close is a factor to calculate if the new values of ascent are very close to each other. If they are, then we stop the
    # function from iterating. If the new values are close for 10 consecutive iterations then we stop the function.
    close = 0
    # we use variable iter to calculate the number of iterations
    iter = 0
    # we initialize two lists to store the final x and function of x values
    X = [x_0]
    # we calculate the value of function for input x
    # we print value of x, value of function and start iterating the ascent
    func = -1*(x_0**2)
    Y = [func]
    derv = -2*x_0
    print("The value of x is ", x_0)
    print("The value of the function is", func)
    print("The value of the derivative is", derv)
    print('starting iterations')
    # we initialize a while loop which iterates until the new values are not close for 10 consecutive iterations
    while(close < 10):
        x_new = x_0 + R*derv
        # increase the iteration values
        func = -1*(x_new**2)
        derv = -2*x_new
        iter+=1
        # if the new function value is very close to old value then we increase the closeness factor
        if(round(x_new,2) == round(x_0,1)):
            close+=1
        else:
            close = 0
        X.append(x_new)
        # we append the function value as y coordinate
        Y.append(func)
        # we replace the old function value with new function value for next iteration
        x_0 = x_new
        # print the x, function of x and gradient value
        print("The new value of the function is", func)
        print("The value of new x is ", x_0)
        print("The value of the gradient is", derv)
    # print the number of iterat
    print("total iterations are:", iter)
    return(X,Y)
    

a,b = grad_asc(0.1, 49)
```
