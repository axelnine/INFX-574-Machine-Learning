```python
x = np.random.randint(low = -50, high = 50, size = 100)
```

```python
print(x)
y = -1*(x**2)  #f(x) = -2x
print(y)
```

```python
def gradient_ascent(R, x_val,stopping_criteria):
    """R : Learning Rate
        x_val: starting value of x
        stopping_criteria: Number of places after the decimal till which the new value and the old val"""
    round_val = 0
    count = 0
    X = list()
    Y = list()
    y = -1 * (x_val**2)
    
    while(round_val < 10):  
        y = -1 * (x_val**2)
        dtv = -2 * x_val  #calculating derivative
        x_new = x_val + R * dtv  #calculate new value of x
        count = count + 1  #increase the iteration count
        
        print("X Value: ", x_val)  
        print("Function Value: ", y)
        print("Gradient Value: ", dtv,"\n")
        
        
        #Epsilon value is 10^-6 
        if(round(x_new,stopping_criteria) == round(x_val,stopping_criteria)):  
            round_val = round_val + 1 
        
        else:
            
            round_val = 0
            
        if(x_val < 0):
            
            X.append(-np.sqrt(-y))
            
        else:
            
            X.append(np.sqrt(-y))
            
        Y.append(y)
        x_val = x_new
        #y = x_new
        
    print("Total Iterations: ", count)
    return(X,Y)
```

```python
a,b = gradient_ascent(0.1, -20,6) 
fig, ax = plt.subplots()
ax.scatter(x,y,color='green')
ax.bar(a,b, color = 'blue')
```

From the above values, we can observe that the total iterations are 89 for the initial X value of -20, learning rate of 0.1 and epsilon is 10^-6

```python
a,b = gradient_ascent(0.1, -40,6)
fig, ax = plt.subplots()
ax.scatter(x,y,color='green')
ax.bar(a,b, color = 'blue')
```

Now, if we change the initial x value to -40, we can see that the number of iterations increase as it takes more iterations to reach the maxima.

```python
a,b = gradient_ascent(0.1, -20,3)
fig, ax = plt.subplots()
ax.scatter(x,y,color='green')
ax.bar(a,b, color = 'blue')
```

Changing the epsilon from 10^-6 to 10^-3, we can see that number of iterations decrease as we increase the epsilon. Hence, decreasing the epsilon value will increase in the number of iterations.

```python
a,b = gradient_ascent(0.05, -20,6)
fig, ax = plt.subplots()
ax.scatter(x,y,color='green')
ax.bar(a,b, color = 'blue')
```

If we decrease the learning rate, the number of iterations increase to 166.

### 1.3 Condition Numbers

-10. Compute the condition number for matrix A in the example above.

```python
from numpy.linalg import inv
def cond_numbers(a):
    a_inv = inv(np.matrix(a))
    sum_a = float(max([sum(a[:,i]) for i in range(len(a))]))
    sum_ainv = float(max([sum(a_inv[:,i]) for i in range(len(a_inv))]))
    cond_val= sum_a * sum_ainv
    print(cond_val)
```

### Pareto Distribution        

```python
a, m = 10, 1
s = (np.random.pareto(a, 100) + 1) * m
import matplotlib.pyplot as plt
count, bins, _ = plt.hist(s, 100, normed=True)
fit = a*m**a / bins**(a+1)
plt.plot(bins, max(count)*fit/max(fit), linewidth=2, color='r')
plt.show()
```
