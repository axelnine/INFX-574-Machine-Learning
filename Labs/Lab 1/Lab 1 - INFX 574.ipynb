### Question 1. 
Ensure that you can repeat the class example with the fliipped-1 matrix and the rotation matrix. Plot the image rotated by a) +45, b) -225, c) an angle of your choice

```python
#Defining the flipped and rotation matrix
A = np.matrix("0 0;0 1;0.5 0.5")

#Define the rotation function and plotting function
#From Prof. Toomet's matrices notebook

def rot(alpha):
    a = alpha*np.pi/180
    return np.matrix([[np.cos(a), -np.sin(a)], [np.sin(a), np.cos(a)]])

def plotFigure(A):
    plt.figure(figsize=(8,8))
    # set figure size
    plt.grid(b=True)
    # show grid
    min_limit = np.min(A) - np.max(A)*2.0
    max_limit = np.max(A) + np.max(A)*2.0
    plt.axes().set_xlim((min_limit, max_limit))
    plt.axes().set_ylim((min_limit, max_limit))
    # set limits (not the best ones but good enough)
    plt.axes().set_aspect('equal','datalim')
    # ensure units are equal along the axes
    plt.plot(A[:,0], A[:,1])
    # plot x, y
    plt.axhline(0, color='black', linestyle = 'dashed')
    plt.axvline(0, color='black',linestyle = 'dashed')
    # draw the axes
    
#This function plot multiple figures in the same plot and is built using the same logic as the plotFigure function
#It also does the dot product for rotating the figure
def plotMultipleFigure(A,B,C,D,angle):
    #rotate angle for each part
    A = np.dot(A,rot(angle))
    B = np.dot(B,rot(angle))
    C = np.dot(C,rot(angle))
    D = np.dot(D,rot(angle))
    plt.figure(figsize=(8,8))
    # set figure size
    plt.grid(b=True)
    # show grid
    # Here we use A since the A called will be the face
    min_limit = np.min(A) - np.max(A)
    max_limit = np.max(A) + np.max(A)
    plt.axes().set_xlim((min_limit, max_limit))
    plt.axes().set_ylim((min_limit, max_limit))
    # set limits (not the best ones but good enough)
    plt.axes().set_aspect('equal','datalim')
    # ensure units are equal along the axes
    plt.plot(A[:,0], A[:,1])
    plt.plot(B[:,0], B[:,1])
    plt.plot(C[:,0], C[:,1])
    plt.plot(D[:,0], D[:,1])
    # plot x, y
    plt.axhline(0, color='black', linestyle = 'dashed')
    plt.axvline(0, color='black',linestyle = 'dashed')
    # draw the axes
```

```python
#Rotated matrices of 45, -224 and 90 degrees
A_45rot = np.dot(A,rot(45))
A_225rot = np.dot(A,rot(-225))
A_90rot = np.dot(A,rot(90))
```

```python
# Actual matrix
plotFigure(A)
```

```python
#45 degree rotated
plotFigure(A_45rot)
```

```python
#-225 degree rotated
plotFigure(A_225rot)
```

```python
#90 degree rotated
plotFigure(A_90rot)
```

## Question 4


Load in the Crazy Hat's image data from the file crazy_hat.tsv. Print the data.

```python
data = pd.read_csv("crazy_hat.tsv", sep="\t")
data
```

```python
#Slice the data according to the groups
outline = data[data.group == 'outline']
leye = data[data.group == 'leye']
reye = data[data.group == 'reye']
mouth = data[data.group == 'mouth']
```

```python
#Convert the series 'x' & 'y' into a matrix
outline_matrix = outline[['x','y']].as_matrix()
leye_matrix = leye[['x','y']].as_matrix()
reye_matrix = reye[['x','y']].as_matrix()
mouth_matrix = mouth[['x','y']].as_matrix()
```

```python
#I have defined dot product in the plotMultipleFigure function itself to avoid writing redundant code

#Plot Crazy Hat with no rotation
plotMultipleFigure(outline_matrix,leye_matrix,reye_matrix,mouth_matrix,0)
```

```python
#Plot Crazy Hat with 135 degree rotation
plotMultipleFigure(outline_matrix,leye_matrix,reye_matrix,mouth_matrix,135)
```

```python
#Plot Crazy Hat with 315 degree rotation
plotMultipleFigure(outline_matrix,leye_matrix,reye_matrix,mouth_matrix,315)
```

```python
#Plot Crazy Hat with -75 degree rotation
plotMultipleFigure(outline_matrix,leye_matrix,reye_matrix,mouth_matrix,-75)
```
